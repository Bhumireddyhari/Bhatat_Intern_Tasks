```python
#importing Libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
df['Postal Code'].fillna(5401, inplace=True)
df.isnull().sum()
```

```python
df.sample(2)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.info()
```

```python
top_cities = df.groupby('City').sum().sort_values('Sales', ascending=False).head(15)

top_cities['Sales'] = top_cities['Sales'].round(2)
top_cities = top_cities.reset_index()

fig = px.bar(top_cities,x='City', y='Sales', title = 'Top 15 cities with respect to sales')
fig.show()
```

```python
top_cities = df.groupby('State').sum().sort_values('Sales', ascending=False).head(15)

top_cities['Sales'] = top_cities['Sales'].round(2)
top_cities = top_cities.reset_index()

fig = px.bar(top_cities,x='State', y='Sales', title = 'Top 15 States with respect to sales')
fig.show()
```

```python
fig = px.pie(df, names='Ship Mode', title='Ship Mode', values=np.ones(df.shape[0]),
             color_discrete_sequence=px.colors.sequential.Aggrnyl_r)
fig.show()
```

```python
 fig = px.pie(df, names='Segment', title='Segment', values=np.ones(df.shape[0]),
             color_discrete_sequence=px.colors.sequential.Aggrnyl_r)
fig.show()
```

```python
fig = px.pie(df, names='Category', title='Category', values=np.ones(df.shape[0]),
             color_discrete_sequence=px.colors.sequential.Aggrnyl_r)
fig.show()
```

```python
top_sales_category = df.groupby('Category').sum()['Sales'].reset_index()
fig = px.pie(top_sales_category,names='Category',values='Sales',
             title = 'Category Withrespect to Sales', color_discrete_sequence=px.colors.sequential.Aggrnyl)
fig.show()
```

```python
fig = px.pie(df, names='Sub-Category', title='Sub-Category', values=np.ones(df.shape[0]),
             color_discrete_sequence=px.colors.sequential.Agsunset)
fig.show()
```

```python
top_sales_region = df.groupby('Region').sum().reset_index()


fig = px.pie(top_sales_region,names='Region',values='Sales'
             ,title='Region wrt Sales',color_discrete_sequence=px.colors.sequential.Aggrnyl)
fig.show()
```

```python
top_sales_segment = df.groupby('Segment').sum().reset_index()


fig = px.pie(top_sales_segment,names='Segment',values='Sales'
             ,title='Segment wrt Sales',color_discrete_sequence=px.colors.sequential.Aggrnyl)
fig.show()
```

```python
plt.figure(figsize=(15,10))

top_cities = df.groupby(['Category','City']).sum().sort_values('Sales',ascending=False).head(40)
top_cities= top_cities[['Sales']].round(2)
top_cities.reset_index(inplace=True)

sns.barplot(x='City',y='Sales',hue='Category',data=top_cities,palette='prism_r')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(15,10))

top_cities= df.groupby(["City","Segment"]).sum().sort_values("Sales",ascending=False).head(40)
top_cities= top_cities[['Sales']].round(2)
top_cities.reset_index(inplace=True)

sns.barplot(x='City',y='Sales',hue='Segment',data=top_cities,palette='prism_r')
plt.xticks(rotation='vertical')
plt.show()
```
